```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
North_Eastern_Railway= pd.read_csv('FEB 2025 Prototype data/NER.csv')
```

```python
# Identify rows where 'Type of A' or 'Type of B' is 'IRS'
irs_rows = North_Eastern_Railway[(North_Eastern_Railway['Type of A'] == 'IRS') | (North_Eastern_Railway['Type of B'] == 'IRS')]

# Display only the first 10 rows
irs_rows.head(3)
```

```python
# Remove rows where 'Type of A' or 'Type of B' is 'IRS'
North_Eastern_Railway = North_Eastern_Railway[(North_Eastern_Railway['Type of A'] != 'IRS') & (North_Eastern_Railway['Type of B'] != 'IRS')]

# Display the cleaned dataset
North_Eastern_Railway.head(10)
```

```python
missing_data =North_Eastern_Railway.isnull().sum()
missing_data
```

```python

# Fill missing values in 'Type of B' with 'TWS' and 'Polling of B' with 100
North_Eastern_Railway['Type of B'] = North_Eastern_Railway['Type of B'].fillna('TWS')
North_Eastern_Railway['Polling of B'] = North_Eastern_Railway['Polling of B'].fillna(100)


Check = North_Eastern_Railway.isnull().sum()
Check
```
